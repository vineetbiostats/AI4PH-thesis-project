```python
import requests
import json

# Define the input text
text = "This is a sample text to be annotated with MESH and OMIM ontologies"

# Define the ontologies to be used
ontologies = "MESH,OMIM"

# Define the BioPortal Annotator API endpoint and parameters
endpoint = "http://data.bioontology.org/annotator"
params = {
    "text": text,
    "ontologies": ontologies,
    "format": "json"
}

# Send a POST request to the API endpoint and get the response as JSON
response = requests.post(endpoint, data=params)
response_json = response.json()

# Create an empty list to store the results
results = []

# Loop through each annotation in the response and extract the relevant information
for annotation in response_json:
    concept = annotation["annotatedClass"]["prefLabel"]
    code = annotation["annotatedClass"]["@id"].split("/")[-1]
    cui = annotation["annotatedClass"]["cui"]
    semantic_type = annotation["annotatedClass"]["semanticType"][0]["name"]
    occurrences = len(annotation["annotations"])
    ontologies = [o["acronym"] for o in annotation["annotatedClass"]["links"]["ontology"]]
    
    # Check if the concept was found in MESH, OMIM or both ontologies
    if "MESH" in ontologies and "OMIM" in ontologies:
        ontology = "both"
    elif "MESH" in ontologies:
        ontology = "MESH"
    else:
        ontology = "OMIM"
    
    # Add the results to the list
    results.append([concept, code, ontology, cui, semantic_type, occurrences])

# Convert the list of results to a pandas DataFrame for easier analysis
output_table = pd.DataFrame(results, columns=["Concept preferred label", "Concept code", "Ontology", "CUI", "Semantic type", "Number of occurrences"])

print(output_table)
```

3, Create an output as a table (.csv or .xlsx file) and design pattern(s) to extract dose (in grams)

To modify the existing Python code to extract dose and duration information and write it to a new output file, we can use regular expressions to search for patterns of numbers and words that correspond to dose and duration information. We can also use the csv module to write the output to a CSV file. Here's the modified code:

```python
import csv
import re
import urllib.request

# Define regular expressions for dose and duration patterns
dose_pattern = r'([0-9]+(\.[0-9]+)?) ?(g|mg|mcg|g\/day|mg\/day|mcg\/day)'
duration_pattern = r'([0-9]+(\.[0-9]+)?) ?(hr|hrs|hour|hours)'

# Load the file from the URL
url = "https://raw.githubusercontent.com/vineetbiostats/AI4PH-thesis-project/main/regex_data.csv"
response = urllib.request.urlopen(url)
input_file = response.read().decode("utf-8")

# Remove lines that start with a number or have length less than 4 characters
output_lines = []
for line in input_file.split("\n"):
    line = line.strip()
    if not line.startswith(tuple(map(str, range(10)))) and len(line) >= 4:
        output_lines.append(line)

# Define headers for output file
headers = ['Substance', 'Dose', 'Duration']

# Initialize list to store output rows
output_rows = [headers]

# Loop over each line in the cleaned input file
for line in output_lines:
    # Split the line into words
    words = line.split()

    # Check if the first word is ceftriaxone
    if words[0].lower() == 'ceftriaxone':
        # Initialize variables to store dose and duration information
        dose = None
        duration = None

        # Loop over each word in the line
        for i in range(len(words)):
            # Search for dose pattern in the word
            match = re.search(dose_pattern, words[i])
            if match:
                # Extract dose information
                dose = match.group()

            # Search for duration pattern in the word
            match = re.search(duration_pattern, words[i])
            if match:
                # Extract duration information
                duration = match.group()

        # If dose and duration information were found, append a new row to output_rows
        if dose and duration:
            output_rows.append(['Ceftriaxone', dose, duration])

# Write the output_rows to a CSV file
with open('output.csv', 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(output_rows)
```

This modified code uses the re module to define regular expressions for dose and duration patterns. It then loops over each line in the cleaned input file and checks if the first word is "ceftriaxone". If so, it initializes variables to store dose and duration information and loops over each word in the line to search for dose and duration patterns. If a match is found, the dose or duration information is extracted and stored in the appropriate variable. If both dose and duration information were found, a new row is appended to output_rows with the substance name ("Ceftriaxone"), dose information, and duration information. Finally, the csv module is used to write the output_rows to a new CSV file called "output.csv".
